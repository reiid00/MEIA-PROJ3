# ROBERTA PreTrained Model

```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
```

```python
MODEL = f'cardiffnlp/twitter-roberta-base-sentiment'
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

Run example in ROBERTA model

```python
encoded_txt = tokenizer(example, return_tensors='pt')
```

```python
output = model(**encoded_txt)
```

```python
output
```

```python
scores = output[0][0].detach().numpy()
```

```python
scores = softmax(scores)
```

```python
scores
```

```python
scores_dict = {
    'roberta_neg' : scores[0],
    'roberta_neu' : scores[1],
    'roberta_pos' : scores[2],
}
```

```python
print(scores_dict)
```

```python
def polarity_scores_roberta(example):
    encoded_txt = tokenizer(example, return_tensors='pt')
    output = model(**encoded_txt)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
    'roberta_neg' : scores[0],
    'roberta_neu' : scores[1],
    'roberta_pos' : scores[2],
    }
    return scores_dict
```

```python
res = {}
df2 = df.head(500)
for i, row in tqdm(df2.iterrows(), total = len(df2)):
    try:
        txt = row['Text']
        rid = row['Id']
        vader_result = sia.polarity_scores(txt)
        vader_result_rename = {}
        for key, value in vader_result.items():
            vader_result_rename[f'vader_{key}'] = value
        roberta_result = polarity_scores_roberta(txt)
        vader_roberta = {** vader_result_rename, ** roberta_result}
        res[rid] = vader_roberta
    except RuntimeError:
        print(f' Broke for id {rid}')
```

```python
resuls_df = pd.DataFrame(res).T
resuls_df = resuls_df.reset_index().rename(columns={'index': 'Id'})
resuls_df = resuls_df.merge(df,how ='left')
```

```python
resuls_df.head()
```

# Compare Scores between Models

```python
fig , axs = plt.subplots (2, 3, figsize=(15, 10))
sns.barplot(data=resuls_df, x='Score',y='vader_pos', ax=axs[0][0])
sns.barplot(data=resuls_df, x='Score',y='vader_neu', ax=axs[0][1])
sns.barplot(data=resuls_df, x='Score',y='vader_neg', ax=axs[0][2])
axs[0][0].set_title('Vader Pos')
axs[0][1].set_title('Vader Neu')
axs[0][2].set_title('Vader Neg')
sns.barplot(data=resuls_df, x='Score',y='roberta_pos', ax=axs[1][0])
sns.barplot(data=resuls_df, x='Score',y='roberta_neu', ax=axs[1][1])
sns.barplot(data=resuls_df, x='Score',y='roberta_neg', ax=axs[1][2])
axs[1][0].set_title('Roberta Pos')
axs[1][1].set_title('Roberta Neu')
axs[1][2].set_title('Roberta Neg')
plt.show
```

```python
sns.pairplot(data=resuls_df,
             vars=['vader_neg', 'vader_neu', 'vader_pos',
                  'roberta_neg', 'roberta_neu', 'roberta_pos'],
            hue='Score',
            palette='tab10')
plt.show()
```
